## Background and Prior Work

Many gamers feel the golden age of video game releases is in the past. Gone are the days of studios releasing polished products that work as intended right out of the box. Instead, many large studios have gotten in the habit of releasing games midway through development with promises to update and fix glaring issues within an update cycle. Not only has this often been done in bad faith, but it has killed the player base of many beloved IPs before the developers can even finish their work. Companies are feeling the effects of constant failure showing the need for a reevaluation of success. This is not only hurting video game companies but also players who are constantly spending money on games only to be disappointed, leaving them to wonder if cheaper Indie games are better.

An example of this failure is Battlefield 2042. This game was expected to be a sure success because it is a big franchise focusing on multiplayer gaming but it turned out to be a disaster. This is primarily due to the game's unprepared launch, marked by inconsistent performance, faulty servers, numerous bugs, and the absence of common features like voice chat and scoreboards, which contributed to its evident mediocrity, resulting in a significant player exodus; the subsequent uncertain future of Battlefield 2042 serves as a substantial wake-up call for the gaming industry. Player counts, the scaling down of lobby sizes, and overall conversation about the game were used to measure its success. This approach is intriguing because it shifts the idea of success from the point of view of companies to their players.  Using revenue and sales metrics alone to quantify success is no longer acceptable, game longevity and player counts over time are now equally if not more important to the success of games and the reputation of companies. It appears that in order to truly succeed AAA companies will have to shift their attention away from sales metrics and instead focus on how to please their audiences or risk losing market share to Indie game developers (King, 2022 )<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1)

One of the main challenges of AAA games is they are excessively large which is linked to their high costs, perpetuating a cycle where games expand to attract attention, requiring larger budgets. To offset these costs, games need to sell more copies, leading to higher prices. The push for $70 games reflects this trend, with titles like Forspoken justifying the price with their open-world nature, potentially overlooking the essence of gaming as an experience rather than a strict correlation of content hours to cost (Polydin, 2023)<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3). In the fiercely competitive landscape of capturing and maintaining audience attention, two categories emerge: the select few that enjoy prolonged attention (e.g., Fortnite, Elden Ring, Skyrim) and those with a fleeting 15 minutes of fame. While AA and Indie games thrive within their niches, AAA games face the demanding task of achieving universal success to avoid being labeled failures, as seen with examples like Battlefield V, which despite selling 7.3 million copies, fell short. The success of AAA games becomes a significant gamble, necessitating not only initial strong sales but also the creation of enduring franchises that engage players deeply (Polydin, 2023)<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3). Instead of large, multi-million dollar studios, many gamers seem to be pivoting towards passion projects produced by indie developers. Battlebit, Baldur’s Gate 3, Risk of Rain 2, etc. were all created on budgets a fraction of the size of AAA companies (Tucker, 2023)<a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4).This shows that smaller game developers are starting to gain and maintain success in the era of video game failures. They have been able to meet players' needs rather than focusing on bigger and better things. This also highlights the importance of player retention ratios and the ratio of new players as a metric for success. Since Indie and AA games may have smaller, more niche communities resulting in smaller player counts, player retention ratios over time and ratio of new players might be better metrics for success than player counts.

## Dataset #4 (SteamCharts)

```python
url = 'https://steamcharts.com/top'
response = requests.get(url)
soup = bs4.BeautifulSoup(response.content, 'html.parser')
```

```python
# scrape data from pages; 25 entries listed per page
data = []
page_count = 4
for i in range(page_count):
    sleep(5)
    if i == 0:
        url = 'https://steamcharts.com/top'
    else:
        url = 'https://steamcharts.com/top' + '/p.' + str(i)
    response = requests.get(url)
    soup = bs4.BeautifulSoup(response.content, 'html.parser')

    # build column data from current page
    table = soup.find('table')
    for row in table.find_all('tr')[1:]:
        cols = row.find_all('td')
        if(len(cols) > 1):
            game_name = cols[1].text.strip()
            current_players = cols[2].text.strip()
            peak_players = cols[4].text.strip()
            hours_played = cols[5].text.strip()
            data.append([game_name, current_players, peak_players, hours_played])
```

```python
df_players = pd.DataFrame(data, columns = ['Game Name', 'Current Players', 'Peak Players', 'Hours Played'])
df_players.head()
```

```python
# write to csv
df_players.to_csv('Data/steamcharts_data.csv', index=False)
```

## Dataset #5 (SteamSpy)

```python
r = requests.get("https://steamspy.com/")
```

```python
soup = bs4.BeautifulSoup(r.text)
# use data from the 'Top by Playtime' tab
body = soup.find_all("tbody")[2].find_all("td")
```

```python
# build column data
names = []
prices = []
owners = []
# for top k games
k = 100
for i in range(k):
    names += [body[(i*8)+1].attrs['data-order']]
    prices += [body[(i*8)+3].text]
    owners += [body[(i*8)+5].text]
```

```python
df_price = pd.DataFrame(data={'Game Name': names, 'Price': prices, 'Number of Owners': owners})
```

```python
# clean price and number_of_owners columns
def fix_price(t):
    if t == 'Free' or t == 'N/A':
        return 0
    return float(t[1:])
def fix_owners(t):
    return [int(t.split('..')[0].replace(',', '').strip()), int(t.split('..')[1].replace(',', '').strip())]
df_price['Price'] = df_price['Price'].apply(fix_price)
df_price['Number of Owners'] = df_price['Number of Owners'].apply(fix_owners)
```

### Section 1 of EDA - Cleaning and Combining Datasets

The code below performs cleaning and combination tasks on three datasets—df_developer, df_players, and df_price—related to gaming data. The goal is to create a consolidated DataFrame for more convenient exploratory data analysis (EDA). The 'Game Name' column is identified as the common key for merging these datasets. To standardize the data, text in the 'Game Name' columns is stripped of non-alphanumeric characters and converted to lowercase. After renaming the 'Game' column in df_developer to 'Game Name' for consistency, the datasets are merged based on this shared column. Duplicate rows based on 'Game Name' are dropped to ensure data integrity. The script also includes functions to extract the maximum value from a range in the 'Number of Owners' column and address special symbols in the 'Employee Count' column. The resulting DataFrame, final_merged_df, is then reset for consistent indexing. Overall, this script prepares the gaming datasets for further analysis by cleaning, standardizing, and combining relevant information from the original datasets.

```python
df_developer = pd.read_csv('Data/developer_data.csv', index_col='Unnamed: 0')
df_players = pd.read_csv('Data/steamcharts_data.csv')
df_price = pd.read_csv('Data/steamspy_data.csv')
```

```python
# Strip all text of anything except uppercase letters, lowercase letters, and numbers
df_developer['Game'] = df_developer['Game'].str.replace('[^a-zA-Z0-9]', '')
df_players['Game Name'] = df_players['Game Name'].str.replace('[^a-zA-Z0-9]', '')
df_price['Game Name'] = df_price['Game Name'].str.replace('[^a-zA-Z0-9]', '')

# Convert the names to lowercase
df_developer['Game'] = df_developer['Game'].str.lower()
df_players['Game Name'] = df_players['Game Name'].str.lower()
df_price['Game Name'] = df_price['Game Name'].str.lower()

# Rename 'Game' column to 'Game Name' in df_developer
df_developer = df_developer.rename(columns={'Game': 'Game Name'})

# Merge DataFrames based on the 'Game Name' column
merged_df = pd.merge(df_players, df_price, on='Game Name', how= 'inner')
merged_df = pd.merge(merged_df, df_developer, on='Game Name', how='inner')

# Drop duplicate rows based on the 'Game Name' column
final_merged_df = merged_df.drop_duplicates(subset='Game Name', keep='first')

# Function to extract the maximum value from the range in 'Number of Owners' column
def extract_max_owners(owners_range):
    owners_range = eval(owners_range)
    if isinstance(owners_range, list):
        return max(owners_range)
    return owners_range

# Apply the function to the 'Number of Owners' column
final_merged_df['Number of Owners'] = final_merged_df['Number of Owners'].apply(extract_max_owners)

# Function to fix two specific instances where employee counts have a special symbol
def fix_employee_count(num):
    if '>' in num:
        return num[1:]
    elif '-' in num:
        return num.split('-')[-1]
    else:
        return num

# Apply the function to the 'Employee Count' column
final_merged_df['Employee_Count'] = final_merged_df['Employee_Count'].apply(fix_employee_count)

# Reset the index for consistency
final_merged_df = final_merged_df.reset_index(drop=True)
```

We analyze the video game dataset, focusing on the distribution and proportions of AAA and non-AAA games over the years. Subsequently, we calculate and visualize the proportions of AAA versus non-AAA games using a stacked bar chart. This concise analysis aims to reveal trends in the gaming industry's evolution, offering insights into how the market landscape has changed over time.

```python
df = pd.read_csv('Data/cleaned_data.csv')
df = df[df['ID'] != 0]
aaa_counts = df['AAA'].value_counts()

# Distribution of games over the years
year_distribution = df['Year_Published'].value_counts().sort_index()

# Preparing data for graph: AAA vs non-AAA over the years
aaa_yearly = df.groupby(['Year_Published', 'AAA']).size().unstack().fillna(0)

# Calculating the proportion of AAA vs non-AAA developers for each year

# Adding a total count for each year
aaa_yearly['Total'] = aaa_yearly.sum(axis=1)

# Calculating the proportion
aaa_yearly['AAA Proportion'] = aaa_yearly[True] / aaa_yearly['Total']
aaa_yearly['Non-AAA Proportion'] = aaa_yearly[False] / aaa_yearly['Total']

# Preparing data for the plot
proportion_data = aaa_yearly[['Non-AAA Proportion','AAA Proportion']]

# Plotting
plt.figure(figsize=(12, 6))
proportion_data.plot(kind='bar', stacked=True, color=['orange', 'skyblue'], ax=plt.gca())
plt.title('Proportion of AAA vs Non-AAA Game Developers Over the Years')
plt.xlabel('Year Published')
plt.ylabel('Proportion of Games')
plt.legend(title='Type', labels=['Non-AAA','AAA'])
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

plt.show()
```

## Section 1 of Analysis - Top 100 and Peak Players Proportions
#### T-test and Augmented Dickey-Fuller Tests

It seems like the proportion of top 100 games has been steadily climbing (with the exception of 2006). We would like to explore if this difference between early and later years is statistically significant.

```python
# Performing a t-test to check if the change in proportion of AAA vs non-AAA developers over the years is significant

# Separate the data into two groups: early years and recent years.
# For this analysis, let's consider 'early years' as before 2015 and 'recent years' as 2015 and after.
early_years = proportion_data[proportion_data.index < 2015]
recent_years = proportion_data[proportion_data.index >= 2015]

# Conducting a t-test on the AAA Proportions
t_stat, p_value = ttest_ind(early_years['AAA Proportion'], recent_years['AAA Proportion'])

t_stat, p_value
```

Our t-test proves that the data before 2015 and the data after 2015 are not part of the same distribution, suggesting there has been a significant change. However, blocking the data like this loses year-by-year change.

As a more advanced analysis, let's try approaching the data as a time series rather than 2 distributions. The Augmented Dickey-Fuller (ADF) test is a statistical test used to determine whether a given time series is stationary. Stationarity is an important property in time series analysis, indicating that the statistical properties of the series (like mean, variance, and autocorrelation) do not change over time. The default assumption of the ADF test is that the time series has a unit root, meaning it is non-stationary. It possesses some form of time-based structure, like a trend or seasonality.

```python
def adf_test(timeseries):
    print ('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print (dfoutput)
# Call the function and run the test

adf_test(aaa_yearly['AAA Proportion'])
```

The ADF test was unable to prove that the time series was stationary (p=.79), which concurs with our original t-test.

It appears that there is a strong statistical difference between the proportion of indie games in the top 100 from before 2015 vs. after (p=.006). But what about actual player count?

```python
# Calculating the total peak players for each year
total_peak_players = df.groupby('Year_Published')['Peak_Players'].sum()

# Calculating the sum of peak players for AAA and non-AAA games each year
aaa_peak_players_sum = df[df['AAA']].groupby('Year_Published')['Peak_Players'].sum()
non_aaa_peak_players_sum = df[~df['AAA']].groupby('Year_Published')['Peak_Players'].sum()

# Calculating the proportion of peak players for AAA and non-AAA games each year
aaa_proportion = aaa_peak_players_sum / total_peak_players
non_aaa_proportion = non_aaa_peak_players_sum / total_peak_players

# Creating a DataFrame for the proportions
peak_players_proportions = pd.DataFrame({
    'AAA Peak Player Proportion': aaa_proportion,
    'Non-AAA Peak Player Proportion': non_aaa_proportion
})

# Plotting the proportions over time for AAA and non-AAA games
plt.figure(figsize=(12, 6))
peak_players_proportions.plot(kind='bar', stacked=True, color=['orange', 'skyblue'], ax=plt.gca())
plt.title('Proportion of Peak Players for AAA vs Non-AAA Games Over the Years')
plt.xlabel('Year Published')
plt.ylabel('Proportion of Peak Players')
plt.legend(title='Game Type', labels=['AAA', 'Non-AAA'])
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

plt.show()
```
